```python
import seaborn
import pandas as pd
import matplotlib.pyplot as plt
import itertools as it
```

ML+feature

```python
def read_cor(cor,method):
    data = pd.read_csv("G:\\ML_result\\DL+ML\\ALL_"+str(cor)+"_"+str(method)+".csv", encoding='utf-8')
    return data

cors=["NDCG", "RJSD", "RRNSS"]
methods=["xgboost","SVM"]

for cor in cors:
    f, axes = plt.subplots(1, 2, figsize=(12, 5 * 1))
    for method,ax in zip(methods,axes):
        data = read_cor(cor,method)
        data.index=['CEFR','CLEC','CLOTH','NES','OSP','RACE']
        d = pd.DataFrame(data)

        seaborn.heatmap(d, annot = True,vmin=0,vmax = 1,square = True,cmap = "Blues",fmt='.3f',ax=ax,annot_kws={"size": 8})
        ax.tick_params(axis='both', which='major')
        ax.set_xlabel("X Label", fontsize=12)
        ax.set_ylabel("Y Label", fontsize=12)
        ax.set_ylabel('Test Sets')
        ax.set_xlabel('Train Sets')
        ax.set_title(str(cor)+"_"+str(method)+"+Feature")
        cbar = ax.collections[0].colorbar

    plt.tight_layout()  
```
