```python
import pylab as plt
import pandas as pd
import seaborn as sns
import itertools as it
```

# Readability formula-level分布图

```python
indexs=['ARI','FleschReadingEase','FleschKincaidGradeLevel','GunningFogIndex','SMOGIndex','ColemanLiauIndex','LIX','RIX']
title=['ARI','FRE','FKGL','GFI','SMOG','CLI','LIX','RIX']
datasets=["CEFR", "CLEC", "CLOTH", "NES", "OSP", "RACE"]
i=-1
for index in indexs:
    i=i+1
    f, axes = plt.subplots(2, 3, figsize=(20, 5 * 2))
    axes = it.chain(*axes)
    for dataset,ax in zip(datasets,axes):
        data = pd.read_excel('index_result\\'+str(dataset)+'_INDEX.xlsx')
        for label in sorted(data["label"].unique()):
            sns.distplot(data[data["label"] == label][index], label=str(label), ax=ax)
        ax.legend()
        ax.set_title("Distribution "+str(title[i])+" Grade per Text by Reading Labels("+str(dataset)+")")
    plt.tight_layout()    
```
